***Cumulative Distribution Function(CDF)***

```python
counts, bin_edges=np.histogram(iris_setosa['PetalLengthCm'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)

#Compute CDF
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.xlabel('PetalLengthCm')
plt.ylabel('Probability')
plt.show()
```

***Clustering***

```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

def cluster(iris):
    x = iris.iloc[:,[0,1,2,3]].values
    wcss=[]
    for i in range(1,11):
        kmeans=KMeans(n_clusters=i, init='k-means++',random_state=30)
        kmeans.fit(x)
        wcss.append(kmeans.inertia_)

    sns.set()
    plt.plot(range(1,11), wcss)
    plt.title('Elbow graph')
    plt.show()
```

I am using the elbow graph to find out the optimum number of clusters

```python
cluster(iris)
```

we can clearly see that the optimum number of cluster 3

```python
kmeans=KMeans(n_clusters=3, init='k-means++', random_state=0)
y=kmeans.fit_predict(x)
```
